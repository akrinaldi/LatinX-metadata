```python
all_Issns = []

for issn in data["Issn"]:
    all_Issns.append(issn)
```

```python
No_results = ['39535','62006','156191','0719367X','991333','1435124','1574180X','1722190','1930826','2500106X','2640473','2763877','7317131','7378831','7408188','7419058','14479567','14711842','15307131','15337979','15590682','18735371','18758649','21506698','24750158']
```

```python
run_again = []
mystery = []

for i in all_Issns:
    for n in No_results:
        if n in i:
            run_again.append(i)
```

```python
print(run_again)
```

```python
for val in run_again:
    for n in No_results:
        if n not in val:
            mystery.append(n)
```

```python
print(mystery)
```

```python
def give_unique(list1):
    unique_ids = []
    for i in list1:
        if i not in unique_ids:
            unique_ids.append(i)
        else:
            continue
    return unique_ids
```

```python
true_mysteries = give_unique(mystery)
```

```python
print(true_mysteries)
```

```python
title_list = []

for ind in data.index:
    key=data["Issn"][ind]
    value=data["Title"][ind]
    for i in true_mysteries:
        if i in key:
            title_list.append(value)
        else:
            continue
        
```

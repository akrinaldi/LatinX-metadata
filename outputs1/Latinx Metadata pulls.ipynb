```python
from urllib import request
```

```python
import time
import json
import os
```

```python
ISSN = ['19384122','13862316']
YEAR_ID = ['2018']
text_sample = ''
URL_PATTERN = 'https://api.elsevier.com/content/search/scopus?&query=issn({})&count=200&date={}&apiKey=ad4e12568b69167a2d25213b52923901&insttoken=0363e8707601cf3e7f27138d2b25f9e4&view=STANDARD'

```

```python
if __name__ == '__main__':
    for issn in ISSN:
        route = "./LatinxMeta/outputs/{i}".format(i = issn)
        if not (os.path.exists(route)):
            os.makedirs(route)
            for y_id in  YEAR_ID:
                if not (os.path.exists(route + "/" + issn + "_" + y_id + ".json")):
                    try:
                        url = URL_PATTERN.format(issn, y_id) 
                        response = request.urlopen(url)   
                        scopus_pull = response.read().decode('utf8')
                        
                        time.sleep(1)
                        
                        full_json = json.loads(scopus_pull)

                        with open(route + "/" + issn + "_" + y_id + ".json", 'w') as scopus:
                            scopus.write(json.dumps(full_json))
                     
                    except Exception as ex:
                        print(str(ex))
                        print(str(url))
                        print(str(response))
                        print("=========================================================")
                        exit()
```
